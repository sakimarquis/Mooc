```python
import pandas as pd
```

```python
# Change the path to the dataset file if needed. 
PATH = '../../data/athlete_events.csv'
```

```python
data = pd.read_csv(PATH)
data.head()
```

__1. How old were the youngest male and female participants of the 1996 Olympics?__

- 16 and 15
- 14 and 12 
- 16 and 12
- 13 and 11

```python
data[(data['Sex'] == 'M') & (data['Year'] == 1996)]['Age'].sort_values().head()
```

```python
data[(data['Sex'] == 'F') & (data['Year'] == 1996)]['Age'].sort_values().head()
```

__2. What was the percentage of male gymnasts among all the male participants of the 2000 Olympics? Round the answer to the first decimal.__

*Hint:* here and further if needed drop duplicated sportsmen to count only unique ones. 

- 0.2
- 1.5 
- 2.5
- 7.7

```python
data['Sport'].unique()
male = data.drop_duplicates('ID')[data['Sex'] == "M"]
100 * male[male['Sport'] == "Gymnastics"]['ID'].count() / male['ID'].count()
```

__3. What are the mean and standard deviation of height for female basketball players participated in the 2000 Olympics? Round the answer to the first decimal.__

- 178.5 and 7.2
- 179.4 and 10
- 180.7 and 6.7
- 182.4 and 9.1 

```python
female_2000 = data[(data['Sex'] == "F") & (data['Year'] == 2000) & (data['Sport'] == "Basketball")]

print(female_2000.drop_duplicates(['ID'])['Height'].mean(),\
female_2000.drop_duplicates(['ID'])['Height'].std())
```

__4. Find a sportsperson participated in the 2002 Olympics, with the highest weight among other participants of the same Olympics. What sport did he or she do?__

- Judo
- Bobsleigh 
- Weightlifting
- Boxing

```python
data[data['Year'] == 2002].sort_values(by = 'Weight', ascending = False).head()
```

__7. Is it true that Switzerland won fewer medals than Serbia at the 2016 Olympics? Do not consider NaN values in _Medal_ column.__ 

- Yes 
- No

```python
print(data[(data['Team'] == "Switzerland") & (data['Year'] == 2016)]['Medal'].dropna().count(), \
data[(data['Team'] == "Serbia") & (data['Year'] == 2016)]['Medal'].dropna().count())
```

__8. What age category did the fewest and the most participants of the 2014 Olympics belong to?__
- [45-55] and [25-35) correspondingly
- [45-55] and [15-25) correspondingly
- [35-45] and [25-35) correspondingly
- [45-55] and [35-45) correspondingly

```python
data2014 = data[data['Year'] == 2014].drop_duplicates(['ID'])
print(data2014[(data2014['Age'] >= 15) & (data2014['Age'] < 25)]['ID'].count(),\
data2014[(data2014['Age'] >= 25) & (data2014['Age'] < 35)]['ID'].count(),\
data2014[(data2014['Age'] >= 35) & (data2014['Age'] <= 45)]['ID'].count(),\
data2014[(data2014['Age'] >= 45) & (data2014['Age'] <= 55)]['ID'].count())
data2014['Age'].hist(color="R", label='[45-55]',  bins = 4)
```

__9. Is it true that there were Summer Olympics held in Lake Placid? Is it true that there were Winter Olympics held in Sankt Moritz?__

- Yes, Yes
- Yes, No
- No, Yes 
- No, No 

```python
print(len(data[(data['Season'] == 'Summer') & (data['City'] == 'Lake Placid')]) > 0,\
len(data[(data['Season'] == 'Winter') & (data['City'] == 'Sankt Moritz')]) > 0)
```
