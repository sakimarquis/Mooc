### 3.数据分割、建立模型、模型学习及评估

#### 1）数据分割

```python
from sklearn.model_selection import train_test_split
#把数据分割成训练集、测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

#### 2）模型建立、学习、评估

```python
import time
from sklearn.linear_model import LinearRegression
#1）模型建立，采用默认参数，或直接 help(LinearRegression)，查看函数的参数设置
model = LinearRegression(normalize=True)
start = time.time()
#2）模型学习 fit
model.fit(X_train, y_train)
#3）模型评估 score
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)
print('elaspe time: {0:.6f}s;\ntrain_score: {1:0.6f};\ntest_score: {2:.6f}'.format(time.time()-start, train_score, test_score))
```

评估指标:

$score=(1-\frac{u}{v})^2$

$u=\sum(y_{pred}-y_{true})^2$

$v=\sum(y_{true}-mean(y_{true}))^2$

1为最优。

#### 3）多项式特征 polynomial
$[a,b]\rightarrow [1,a,b,a^2,b^2,ab]$



```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
#利用pipeline将多个阶段的处理单元组合成一个整体模型
def polynomial_model(degree=1):
    polynomial_features = PolynomialFeatures(degree=degree,
                                             include_bias=False)
    linear_regression = LinearRegression(normalize=True)
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])
    return pipeline
#1）建立模型
model = polynomial_model(degree=2)

start = time.time()
#模型学习数据
model.fit(X_train, y_train)
#评估模型在训练和测试数据上的效果
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)
print('elaspe time: {0:.6f};\ntrain_score: {1:0.6f};\ntest_score: {2:.6f}'.format(time.time()-start, train_score, test_score))
```

#### 4）多项式不同次数的影响


```python
from sklearn.model_selection import ShuffleSplit
#打散后数据分割 = 多次train_test_split
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
degrees = [1,2,3]
title = 'score of degrees: {}'.format(degrees)
plt.figure(figsize=(10,8))
#用cv提供的数据分割，对每个degree的模型测试效果
for i in range(len(degrees)):
    poly=polynomial_model(degrees[i])
    train_scores=[]
    test_scores=[]
    for train,test in cv.split(X):
        poly.fit(X[train],y[train])
        train_scores.append(poly.score(X[train],y[train]))
        test_scores.append(poly.score(X[test],y[test]))
    #print(test_scores)
    train_mean=np.mean(train_scores)
    train_std=np.std(train_scores)
    test_mean=np.mean(test_scores)
    test_std=np.std(test_scores)
    print('degree: {:d}, train score: {:.6f}, test score: {:.6f}'.format(degrees[i],train_mean,test_mean))
    plt.bar([2*(i),2*(i)+1],[train_mean,test_mean],width=0.8,yerr=[train_std,test_std],color=np.random.rand(1,3),label='degree %d'%i)
plt.title(title)
plt.legend()
plt.xticks(np.arange(len(degrees)*2),['train','test']*len(degrees))
plt.ylabel('$R^2$')
plt.show()

```
