# Chapter 8 - Comparing Gaussian means
## 8.1 One-sample comparison


$$ \delta \sim \text{Cauchy} (0, 1)$$
$$ \sigma \sim \text{Cauchy} (0, 1)_{\mathcal I(0,∞)}$$
$$ \mu = \delta\sigma $$
$$ x_{i} \sim \text{Gaussian}(\mu,1/\sigma^2)$$

```python
# Read data Dr. Smith
winter = np.array([-0.05,0.41,0.17,-0.13,0.00,-0.05,0.00,
                   0.17,0.29,0.04,0.21,0.08,0.37,0.17,0.08,-0.04,
                   -0.04,0.04,-0.13,-0.12,0.04,0.21,0.17,0.17,0.17,
                   0.33,0.04,0.04,0.04,0.00,0.21,0.13,0.25,-0.05,
                   0.29,0.42,-0.05,0.12,0.04,0.25,0.12])

summer = np.array([0.00,0.38,-0.12,0.12,0.25,0.12,0.13,
                   0.37,0.00,0.50,0.00,0.00,-0.13,-0.37,-0.25,-0.12,
                   0.50,0.25,0.13,0.25,0.25,0.38,0.25,0.12,0.00,0.00,
                   0.00,0.00,0.25,0.13,-0.25,-0.38,-0.13,-0.25,0.00,
                   0.00,-0.12,0.25,0.00,0.50,0.00])

x = winter - summer  # allowed because it is a within-subjects design
delta_obs = x / np.std(x) # effect size
```

```python
with pm.Model() as model1:
    delta = pm.Cauchy("delta", alpha=0.0, beta=1.0)
    sigma = pm.HalfCauchy("sigma", beta=1.0)
    mu = delta * sigma
    xi = pm.Normal("xi", mu=mu, sd=sigma, observed=delta_obs)

    trace1 = pm.sample(3000)
    az.plot_trace(trace1)
```

```python
az.summary(trace1)
```

```python
def display_delta(trace, x_range):
    """
    Use Savage– Dickey method to estimate Bayesian factor
    BFs based on density estimation (using kernel smoothing instead of spline)
    """
    # estimate posterior
    posterior_kde = stats.kde.gaussian_kde(trace["delta"][:])
    # plot distribution function
    plt.plot(x, posterior_kde(x), "--", lw=2.5, alpha=0.6, label="Posterior")  
    plt.plot(x, stats.cauchy.pdf(x), "r-", lw=2.5, alpha=0.6, label="Prior")
    
    posterior_0 = posterior_kde(0)  # this gives the pdf at point delta = 0
    prior_0 = stats.cauchy.pdf(0)  # height of order-restricted prior at delta = 0
    BF01 = posterior_0 / prior_0 # the Savage– Dickey method
    print("the Bayes Factor is %.5f" % (BF01))
    
    # plot two points
    plt.plot([0, 0], [posterior_0, prior_0], "k-", lw=1.5)
    plt.plot([0, 0], [posterior_0, prior_0], "ko", alpha=1)
    plt.xlabel("Delta")
    plt.ylabel("Density")
    plt.legend(loc="upper left")
    plt.show()

x = np.linspace(-3, 3, 100)
display_delta(trace1, x)
```

## 8.2 Order-restricted one-sample comparison


$$ \delta \sim \text{Cauchy} (0, 1)_{\mathcal I(-∞,0)}$$
$$ \sigma \sim \text{Cauchy} (0, 1)_{\mathcal I(0,∞)}$$
$$ \mu = \delta\sigma $$
$$ x_{i} \sim \text{Gaussian}(\mu,1/\sigma^2)$$

```python
# Read data Dr. Smith
winter = np.array([-0.05,0.41,0.17,-0.13,0.00,-0.05,0.00,
                   0.17,0.29,0.04,0.21,0.08,0.37,0.17,0.08,-0.04,
                   -0.04,0.04,-0.13,-0.12,0.04,0.21,0.17,0.17,0.17,
                   0.33,0.04,0.04,0.04,0.00,0.21,0.13,0.25,-0.05,
                   0.29,0.42,-0.05,0.12,0.04,0.25,0.12])

summer = np.array([0.00,0.38,-0.12,0.12,0.25,0.12,0.13,
                   0.37,0.00,0.50,0.00,0.00,-0.13,-0.37,-0.25,-0.12,
                   0.50,0.25,0.13,0.25,0.25,0.38,0.25,0.12,0.00,0.00,
                   0.00,0.00,0.25,0.13,-0.25,-0.38,-0.13,-0.25,0.00,
                   0.00,-0.12,0.25,0.00,0.50,0.00])

x = winter - summer  # allowed because it is a within-subjects design
delta_obs = x / np.std(x) # effect size
```

```python
with pm.Model() as model2:
    delta_pos = pm.HalfCauchy("delta_pos", beta=1.0)
    delta = pm.Deterministic("delta", -delta_pos)
    sigma = pm.HalfCauchy("sigma", beta=1.0)
    mu = delta * sigma
    xi = pm.Normal("xi", mu=mu, sd=sigma, observed=delta_obs)

    trace2 = pm.sample(3000, tune=2000)
    az.plot_trace(trace2, var_names=["delta","sigma"])
```

```python
x = np.linspace(-3, 0, 100) # truncated
display_delta(trace2, x)
```

## 8.3 Two-sample comparison


$$ \delta \sim \text{Cauchy} (0, 1)$$
$$ \mu \sim \text{Cauchy} (0, 1)$$
$$ \sigma \sim \text{Cauchy} (0, 1)_{\mathcal I(0,∞)}$$
$$ \alpha = \delta\sigma $$
$$ x_{i} \sim \text{Gaussian}(\mu+\frac{\alpha}{2},1/\sigma^2)$$
$$ y_{i} \sim \text{Gaussian}(\mu-\frac{\alpha}{2},1/\sigma^2)$$


```python
# Read data
x = np.array([70, 80, 79, 83, 77, 75, 84, 78, 75, 75, 
              78, 82, 74, 81, 72, 70, 75, 72, 76, 77])

y = np.array([56, 80, 63, 62, 67, 71, 68, 76, 79, 67, 
              76, 74, 67, 70, 62, 65, 72, 72, 69, 71])

n_x, n_y = len(x), len(y)
print(n_x, n_y)

# Rescale, use x group
y = (y - np.mean(x)) / np.std(x)
x = (x - np.mean(x)) / np.std(x)
```

```python
with pm.Model() as model3:
    delta = pm.Cauchy("delta", alpha=0, beta=1)
    mu = pm.Cauchy("mu", alpha=0, beta=1)
    sigma = pm.HalfCauchy("sigma", beta=1)

    alpha = delta * sigma
    xi = pm.Normal("xi", mu=mu + alpha / 2, sd=sigma, observed=x)
    yi = pm.Normal("yi", mu=mu - alpha / 2, sd=sigma, observed=y)

    trace3 = pm.sample(3000)
    az.plot_trace(trace3)
```

```python
x = np.linspace(-3, 3, 100)
display_delta(trace3, x)
```
